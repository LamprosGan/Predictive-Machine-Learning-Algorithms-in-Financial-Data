```python
lob = data.at['TPEIR','B']
```

```python

lob = lob.drop_duplicates()
lob = lob.dropna()
lob = lob.sort_values(by=['Publication Timestamp'])
lob = lob.set_index('Publication Timestamp')
columns_to_keep = [
    'bid_price_1', 'bid_size_1',
    'ask_price_1', 'ask_size_1',  
    'bid_price_2', 'bid_size_2',  
    'ask_price_2', 'ask_size_2',  
    'bid_price_3', 'bid_size_3', 
    'ask_price_3', 'ask_size_3',  
    'bid_price_4', 'bid_size_4',  
    'ask_price_4', 'ask_size_4',  
    'bid_price_5', 'bid_size_5', 
    'ask_price_5', 'ask_size_5',  
    'bid_price_6', 'bid_size_6',  
    'ask_price_6', 'ask_size_6',  
    'bid_price_7', 'bid_size_7', 
    'ask_price_7', 'ask_size_7',  
    'bid_price_8', 'bid_size_8', 
    'ask_price_8', 'ask_size_8', 
    'bid_price_9', 'bid_size_9', 
    'ask_price_9', 'ask_size_9',  
    'bid_price_10', 'bid_size_10',  
    'ask_price_10', 'ask_size_10', 
]

lob = lob[columns_to_keep]

# Convert index to datetime if not already
lob.index = pd.to_datetime(lob.index)

# Filter for specific trading hours
lob = lob.between_time('10:30', '17:00')
```

```python
# 1. Convert prices to integers by scaling with 10,000
price_columns = [col for col in lob.columns if "price" in col]
size_columns = [col for col in lob.columns if "size" in col]

lob[price_columns] = (lob[price_columns] * 10000).astype(int)
lob[size_columns] = lob[size_columns].astype(int)

# 2. Reorder columns to LOBSTER format.
num_levels = 10
lobster_columns = []
for i in range(1, num_levels+1):
    ask_price_col = f"ask_price_{i}"
    ask_size_col = f"ask_size_{i}"
    bid_price_col = f"bid_price_{i}"
    bid_size_col = f"bid_size_{i}"
    # Append columns in LOBSTER order for this level
    lobster_columns += [ask_price_col, ask_size_col, bid_price_col, bid_size_col]

# Reindex lob with these new columns to match the LOBSTER structure
df_lobster_format = lob[lobster_columns]
```

```python
# This groups data by the date of the index
daily_groups = df_lobster_format.groupby(df_lobster_format.index.date)

ticker = "TPEIR"
num_levels = 10 
```

```python
for day, group in daily_groups:
# Print the date and the number of rows for this day's data
    print(f"Date: {day}, Length: {len(group)} rows")

    # Proceed to save the file as before
    day_str = day.strftime("%Y-%m-%d")
    start_time = group.index.min().strftime("%H:%M:%S")
    end_time = group.index.max().strftime("%H:%M:%S")

    filename = f"{ticker}_{day_str}_{start_time}_{end_time}_orderbook_{num_levels}.csv"
    filepath = f"./days/{filename}"
    group.to_csv(filepath, index=False, header=False, encoding='utf-8')

```
